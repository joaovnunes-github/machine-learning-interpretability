# 01 — Minimal Pipeline
Load data → basic cleaning → train Decision Tree → evaluate.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from sklearn.tree import DecisionTreeClassifier

from src.data import load_data, split_data
from src.evaluate import evaluate
```

```python
# Drops DEATH_EVENT from features and returns it separately as the prediction target
X, y = load_data('../data/heart_failure_clinical_records_dataset.csv')
print(X.shape)
X.head()
```

```python
# Dataset has no missing values, so no imputation is needed at this stage
print(X.isnull().sum())

# 203 survived (0) vs 96 deceased (1) — moderately imbalanced, F1 is a more honest metric than accuracy
print('\nTarget distribution:')
print(y.value_counts())
```

```python
# Stratified split preserves the 68/32 class ratio in both train and test sets
X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
# Baseline Decision Tree with no depth limit — likely overfits, but establishes a reference point
# random_state=42 ensures reproducibility across runs
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
results = evaluate(dt, X_test, y_test, model_name='Decision Tree (baseline)')
```
