# 02 — Model Comparison
Train Decision Tree, KNN, and Naive Bayes through the preprocessing pipeline and compare metrics.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd

from src.data import load_data, split_data
from src.models import train_decision_tree, train_knn, train_naive_bayes
from src.evaluate import evaluate
```

```python
X, y = load_data('../data/heart_failure_clinical_records_dataset.csv')
X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
# Decision Tree — scaling has no effect on splits, result should match Step 1 baseline
dt = train_decision_tree(X_train, y_train)
dt_results = evaluate(dt, X_test, y_test, model_name='Decision Tree')
```

```python
# KNN — scaling is critical here; raw feature ranges vary wildly
knn = train_knn(X_train, y_train, k=5)
knn_results = evaluate(knn, X_test, y_test, model_name='KNN (k=5)')
```

```python
# Naive Bayes — probabilistic model, no tuning needed at this stage
nb = train_naive_bayes(X_train, y_train)
nb_results = evaluate(nb, X_test, y_test, model_name='Naive Bayes')
```

```python
# Comparative metrics table
results = [
    {'Model': 'Decision Tree', **dt_results},
    {'Model': 'KNN (k=5)',     **knn_results},
    {'Model': 'Naive Bayes',   **nb_results},
]

df_results = pd.DataFrame(results).set_index('Model')
df_results.columns = ['Accuracy', 'F1 (weighted)']
df_results = df_results.round(4)
df_results
```
